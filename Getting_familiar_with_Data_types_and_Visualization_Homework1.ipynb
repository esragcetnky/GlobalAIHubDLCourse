## 2.1 CSV

*2.1.1 Reading Csv file*
* Source of dataset : https://www.kaggle.com/alopez247/pokemon

```python
csvFile=pd.read_csv('drive/MyDrive/GitHub/GlobalAI/data/pokemon.csv')
```

*1.2 Visualize Csv file*

```python
# shape, dimension and size of csv file
print('Shape :' +str(csvFile.shape) + '\nDimension :'+ str(csvFile.ndim) + '\nSize :' + str(csvFile.size))
```

```python
# printing columns 
csvFile.columns
```

```python
# created figure with 3 ax in it
fig, (ax0,ax1)=plt.subplots(figsize=(20,15), # size of figure
                            ncols=1,         
                            nrows=2)

# ax0 is a scatter plot
plot0=ax0.scatter(csvFile['Attack'],
                csvFile['Defense'],
                c=csvFile['isLegendary'],
                cmap='winter')

# giving ax0 a legend with title
ax0.legend(*plot0.legend_elements(),
          title="Is Legendary?")

ax0.set(title="Relation Between Attack, Defense, Legendary",
       xlabel='Attack',
       ylabel='Defense')

# setting limits of x and y axises
ax0.set_xlim([0,175])
ax0.set_ylim([0,250])

#ax1 is a scatter plot 
plot1=ax1.scatter(csvFile['Speed'],
                csvFile['HP'],
                c=csvFile['isLegendary'],
                cmap='Set1')

# giving ax1 a legend with title
ax1.legend(*plot1.legend_elements(),
          title="Is Legendary?")

# in this code we are giving ax1 title and labels
ax1.set(title="Relation Between Speed, HP, Legendary",
       xlabel='Speed',
       ylabel='HP')

# setting limits of x and y axises
ax1.set_xlim([0,170])
ax1.set_ylim([0,270])

# showing the finale figure
fig.show()
```

## 2.2 TSV

* Source of Tsv file : https://www.kaggle.com/Cornell-University/movie-dialog-corpus

```python
# only difference between csv and tsv file is csv seperate columns with comma 
# tsv is seperating them with tab so we can read tsv file with pandas.read_csv function
tsvFile = pd.read_csv('drive/MyDrive//GitHub/GlobalAI/data/movie_titles_metadata.tsv',
                      sep='\t',
                      names=['sample number','movie name','year','IMDB ratings','# Imdb votes','genre of the movie'])
```

```python
# columns of tsv file
tsvFile.columns
```

```python
# creating a figure which size is equal to (10,10)
fig, ax=plt.subplots(figsize=(10,10))
# plotting a histogram 
plot = ax.hist(tsvFile['IMDB ratings'],bins=15,color='red')
# setting title and labels of ax
ax.set(title="IMDB ratings",
        xlabel='Ratings',
        ylabel='Number of movies')
# plotting figure
fig.show()
```

## 2.5 ODS

*2.5.1 Reading ODS file*
* Source of ODS file : https://file-examples.com/index.php/sample-documents-download/sample-ods-download/

```python
# installing pandas_ods_reader library
!pip install pandas_ods_reader
```

```python
from pandas_ods_reader import read_ods
odsFile = read_ods('drive/MyDrive/GitHub/GlobalAI/data/file_example_ODS_100.ods',1)
```

*2.5.2 Visualizing ODS file*

```python
print('Shape :' +str(odsFile.shape) + '\nDimension :'+ str(odsFile.ndim) + '\nSize :' + str(odsFile.size))
```

```python
odsFile
```

```python
# creating figure
def label_function(val):
    return f'{val / 100 * len(odsFile):.0f}\n{val:.0f}%'


fig, (ax0,ax1) =plt.subplots(figsize=(20,10),
                                   nrows=1,
                                   ncols=2)

odsFile.groupby('Country').size().plot(kind='pie', 
                                       autopct=label_function, 
                                       textprops={'fontsize': 20},
                                       colors=['tomato', 'gold', 'skyblue'], 
                                       ax=ax0)
odsFile.groupby('Gender').size().plot(kind='pie', 
                                      autopct=label_function, 
                                      textprops={'fontsize': 20},
                                      colors=['violet', 'lime'],
                                      ax=ax1)

ax0.set_ylabel('Per country', size=22)
ax1.set_ylabel('Per gender', size=22)

# showing the figure
fig.show()
```
